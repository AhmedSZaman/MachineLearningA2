```python
import gym
import numpy as np

# Define the Q-learning agent
class QLearningAgent:
    def __init__(self, env, learning_rate=0.1, discount_factor=0.99, epsilon=0.1):
        self.env = env
        self.learning_rate = learning_rate
        self.discount_factor = discount_factor
        self.epsilon = epsilon
        self.q_table = np.zeros((env.observation_space.shape[0], env.observation_space.shape[1], env.action_space.n))
    def choose_action(self, state):
        state = tuple(state.astype(int))  # Convert state to tuple of integers
        actions = np.argmax(self.q_table[state], axis=-1)
        return actions.astype(int)

    
    def update_q_table(self, state, action, reward, next_state, done):
        state = state.astype(int)
        next_state = next_state.astype(int)
        q_value = self.q_table[state[0], state[1], action]
        if done:
            q_value += self.learning_rate * (reward - q_value)
        else:
            max_q_value = np.max(self.q_table[next_state[0], next_state[1]])
            q_value += self.learning_rate * (reward + self.discount_factor * max_q_value - q_value)
        self.q_table[state[0], state[1], action] = q_value

```

```python

# Create the environment and agent
agent = QLearningAgent(env)

# Q-learning training loop
num_episodes = 1000

for episode in range(num_episodes):
    state = env.reset()
    done = False

    while not done:
        action = agent.choose_action(state)
        next_state, reward, done, _ = env.step(action)
        agent.update_q_table(state, action, reward, next_state, done)
        state = next_state

    if (episode + 1) % 100 == 0:
        print(f"Episode {episode + 1}/{num_episodes} completed")

# Evaluate the trained agent
total_reward = 0
num_eval_episodes = 10

for _ in range(num_eval_episodes):
    state = env.reset()
    done = False

    while not done:
        action = np.argmax(agent.q_table[state])
        state, reward, done, _ = env.step(action)
        total_reward += reward

average_reward = total_reward / num_eval_episodes
print("Average reward:", average_reward)
```
